String Methods
Write a program that:
Asks the user to enter a text
Prints the text in uppercase
Prints the text in lowercase
Prints the number of characters in the text

```python
text = input("Enter a text: ")
upperResult = text.upper()
lowerResult = text.lower()
lenResult = len(text)
print(upperResult, lowerResult, lenResult, sep="\n")
```

 Write a Python Program  that takes a string as input and print the string in reverse order.

```python
s = input("Enter a word: ")
result = s[::-1]
print(result)
```
