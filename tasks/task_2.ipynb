# **Task 1: Compare Two Numbers:**
* Take two numbers as input from the user.

* Use an if condition to print the larger number.

* If the numbers are equal, print "Both are equal".

Example:

Input: 10, 20

Output: 20 is greater

---

```python
num1 = int(input())
num2 = int(input())

if num1 > num2:
    print(f"{num1} is greater")
elif num2 > num1:
    print(f"{num2} is greater")
else:
    print(f"Both are equal")
    
print(f"{num1} is num1")
print(f"{num2} is num2")
```

```python
num1 = int(input())
num2 = int(input())

match num1, num2:
    case g if num1 > num2:
        print(f"{num1} is greater")
    case g if num1 < num2:
        print(f"{num2} is greater")
    case _:
        print("Both are equal")
        
print(f"{num1} is num1")
print(f"{num2} is num2")
```

# **Task 2: Vowel or Consonant**

Ask the user to enter a single letter.

Use an if condition to check if it is a vowel (a, e, i, o, u).

Print "Vowel" if it is a vowel, otherwise print "Consonant".

Example:

Input: e

Output: Vowel

---

```python
letter = input().lower()
vowel = ["a", "e", "i", "o", "u"]

if letter in vowel:
    print(f'This letter "{letter}" is vowel')
else:
    print(f'This letter "{letter}" is Consonant')
```

```python
letter = input().lower()

match letter:
    case "a" | "e" | "i" | "o" | "u":
        print(f'This letter "{letter}" is vowel')
    case _:
        print(f'This letter "{letter}" is Consonant')
```

# **Task 3: Password Check**

Set a fixed password in the program, e.g., "Open123".

Ask the user to enter a password.

Use an if condition to check if it matches the fixed password.

Print "Access Granted" if correct, otherwise "Access Denied".

Example:

Input: Open123

Output: Access Granted


---

```python
enterPass = input()
password = "Open123"

if enterPass == password:
    print("Access Granted")
else:
    print("Access Denied")
```

```python
enterPass = input()

match enterPass:
    case "Open123":
        print("Access Granted")
    case _:
        print("Access Denied")
```

# **Task 4: Password Check : Even or Odd**

Write a program that:

Asks the user to enter a number.

If the number is even, print "Even Number".

Otherwise, print "Odd Number".

---

```python
num = int(input())

if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
num = int(input())

match num:
    case n if n % 2 == 0:
        print(f"{num} is even")
    case _:
        print("Odd")
```

# **Task 5: Grade Evaluation**

Write a program that:

Asks the user to enter their score.

Prints:

"Excellent" if the score is 90 or above.

"Very Good" if the score is between 75 and 89.

"Good" if the score is between 60 and 74.

"Fail" if the score is less than 60.

---

```python
score = int(input())
print(f"My Score: {score}")

if score >= 90:
    print("Excellent")
elif score >= 75 and score <= 89:
    print("Very Good")
elif score >= 60 and score <= 74:
    print("Good")
elif score < 60:
    print("Fail")
else:
    ("Wrong Value")
    
```

```python
score = int(input())
print(f"My Score: {score}")

match score:
    case s if s >= 90:
        print("Excellent")
    case s if s >= 75 or s <= 89:
        print("Very Good")
    case s if s >= 60 or s <= 74:
        print("Good")
    case s if s < 60:
        print("Fail")
    case _:
        print("Wrong Value")
```

# **Task 6: Number Comparison**

Write a program that:

Asks the user to enter a number.

Prints:

"Zero" if the number is 0.

"Positive" if the number is greater than 0.

"Negative" if the number is less than 0.

---

```python
num = int(input())

if num > 0:
    print(f"{num} is Positive")
elif num == 0:
    print(f"{num} is Zero")
elif num < 0:
    print(f"{num} is Negative")
else:
    print("Wrong Value")
```

```python
num = int(input())

match num:
    case n if n > 0:
        print(f"{num} is Positive")
    case n if n == 0:
        print(f"{num} is Zero")
    case n if n < 0:
        print(f"{num} is Negative")
    case _:
        print("Wrong Value")
```
